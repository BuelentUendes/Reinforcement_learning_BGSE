```python
N=100
state_space = np.arange(0,N)
q_low = 0.51
c_low = 0
q_high = 0.6
c_high = 0.01
gamma = 0.9

def reward_fnctn(x, N, c):
    return - ((x/N)**2) - c
```

## Problem 1:

+ X is a set of states
+ A is a set of actions
+ X' is the set of next respective states of X


The first column is the state at t, and all the other columns are the possibilities at t+1.

_NB:_ element **(0,0)** of P is **P(observe 0 | previous state was 0, pi(0))**, and element **(3,5)** of P is **P(observe 5 | previous state was 3, pi(3))** 

We have 3 cases:
+ $X_{t+1}$ = $X_{t}$ when P(St=1, It=1) + P(St=0, It=0) = P(St=1) * P(It=1) + P(St=0) * P(It=0) = $q(at)*(0.5) + (1 - q(at))*(0.5)$
+ $X_{t+1}$ = $X_{t}+1$ when P(St=1, It=1) = P(St=0) * P(It=1) = $ (1 - q(at))*(0.5)$
+ $X_{t+1}$ = $X_{t}-1$ when P(St=1, It=0) = P(St=1) * P(It=0) = $q(at)*(0.5)$

, assuming independance.

### Lazy Policy:

```python
def prob_constructor(N, q):
    P_lazy = np.zeros((N,N))
    
    for i in range(N):
        for j in range(N):
            
            #Only take cases that are either x-1, x, or x+1, so the others remain 0 since by structure
            #they have all probabilities = 0.
            
            #Xt+1 = Xt - 1
            if j == i-1: 
                P_lazy[i, j] = 0.5*q
            
            #Xt+1 = Xt
            elif j == i: 
                P_lazy[i, j] = 0.5*q + 0.5*(1-q)
                
            #Xt+1 = Xt + 1
            elif j == i+1: 
                P_lazy[i, j] = 0.5*(1-q)

    #Condition for truncation: 
    #so here we add the probability of P(St=1,It=0)+P(St=1, It=1)+P(St01, It=0)
    #because the probability of moving from -1 to 0 is just the prob of staying at 0  
    P_lazy[0, 0] = 0.5*q + 0.5*q + 0.5*(1-q)
    
    P_lazy[99, 99] = 0.5*(1-q) + 0.5*q + 0.5*(1-q)
    
    return P_lazy


def reward(N,c):
    rewards = []
    for i in range(N):
        rewards.append(reward_fnctn(i, N, c)) 
    return np.array(rewards)
```

```python
P_lazy = prob_constructor(N, q_low)
r_lazy = reward(N, c_low) 
V_lazy = np.dot(np.linalg.inv(np.identity(N) - gamma*P_lazy), r_lazy)
```

### Aggressive Policy:

```python
def aggr_prob_constructor(N, q_low, q_high):
    P_aggr = np.zeros((N,N))  
    for i in range(N):
        for j in range(N):
            
            #use low service rate for short queues:
            if i < 50:
                if j == i-1: 
                    P_aggr[i, j] = 0.5*q_low
                elif j == i: 
                    P_aggr[i, j] = 0.5*q_low + 0.5*(1-q_low)
                elif j == i+1: 
                    P_aggr[i, j] = 0.5*(1-q_low)
            
            #use high service rate for long queues:
            else:
                if j == i-1: 
                    P_aggr[i, j] = 0.5*q_high
                elif j == i: 
                    P_aggr[i, j] = 0.5*q_high + 0.5*(1-q_high)
                elif j == i+1: 
                    P_aggr[i, j] = 0.5*(1-q_high)

                    
    #Condition for truncation: 
    P_aggr[0, 0] = 0.5*q_low + 0.5*q_low + 0.5*(1-q_low)
    
    P_aggr[99, 99] = 0.5*(1-q_high) + 0.5*q_high + 0.5*(1-q_high)
            
    return P_aggr


def aggr_reward(N, c_low, c_high):
    rewards = []
    for i in range(N):
        if i < 50:
            rewards.append(reward_fnctn(i, N, c_low))
        else:
            rewards.append(reward_fnctn(i, N, c_high))
        
    return np.array(rewards)
```

```python
P_aggr = aggr_prob_constructor(N, q_low, q_high)
r_aggr = aggr_reward(N, c_low, c_high) 
V_aggr = np.dot(np.linalg.inv(np.identity(N) - gamma*P_aggr), r_aggr)
```

### Value Iteration:

```python
def value_iteration(V0, N, tol, gamma, c_low, c_high, q_low, q_high):
    t_start = process_time() 
    
    V_old = np.array(V0)
    V_10 = V_20 = V_50 = V_100 = np.empty(N)
    r_low = reward(N, c_low)
    P_low = prob_constructor(N, q_low) 
    r_high = reward(N, c_high)
    P_high = prob_constructor(N, q_high)
    
    V_new_low = r_low + np.dot((gamma*P_low), V_old)
    V_new_high = r_high + np.dot((gamma*P_high), V_old)
    V_new = np.maximum(V_new_low, V_new_high)
    
    i=1
    while (np.absolute(V_new - V_old) > tol).all():
        V_old = np.array(V_new)
        V_new_low = r_low + np.dot((gamma*P_low), V_old)
        V_new_high = r_high + np.dot((gamma*P_high), V_old)
        V_new = np.array(np.maximum(V_new_low, V_new_high))
        
        #Storing the necessary iterations
        if i == 10:
            V_10 = np.array(V_new)
        elif i == 20:
            V_20 = np.array(V_new)
        elif i == 50:
            V_50 = np.array(V_new)           
        elif i == 100:
            V_100 = np.array(V_new)
            t_stop = process_time()
        i+=1

    #To extract the policy:
    policies = []
    for j in range(N):
        if V_new_low[j]>V_new_high[j]:
            policies.append(0)
        else:
            policies.append(1)

    time_100_iter = t_stop - t_start
    return V_new, V_10, V_20, V_50, V_100, i, time_100_iter, policies
```

```python
N = 100
tol = 1e-7

#choose V0 as just rewards using low action rewards
V0 = reward(N, c_low)

V_value, V_10, V_20, V_50, V_100, i_value, time_100, policies_value = value_iteration(V0, N, tol, gamma, c_low, c_high, q_low, q_high)
```

```python
figure(figsize=(10, 6))
plt.plot(range(N), V_value, label = 'Optimal')
plt.plot(range(N), V_10, label = '10 iterations')
plt.plot(range(N), V_20, label = '20 iterations')
plt.plot(range(N), V_50, label = '50 iterations')
plt.plot(range(N), V_100, label = '100 iterations')
plt.title('Value Iteration Algorithm')
plt.legend()
```
