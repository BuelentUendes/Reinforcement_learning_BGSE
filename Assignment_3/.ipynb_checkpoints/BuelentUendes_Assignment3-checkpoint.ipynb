# Coding up the functions 

```python
# Reward function

def reward_function(x, a,N):
    if a == 0:
        return -(x/N)**2 
    else:
        return -(x/N)**2 - 0.01
    
#Transition function

def transition_function(N, a):
    #First get the increment for the increase in que lenght
    I = np.random.binomial(1,p)
    
    if a == 1:
        S = np.random.binomial(1, q_high)
    else:
        S = np.random.binomial(1, q_low)
    
    return min(N-1, max(x+I-S,0)) 


# Reward function 

def reward_policy_function(N, lazy = True):
    output = []
    if lazy == True:
        for i in range(0,N):
            output.append(reward_function(i, 0,N))
    else:
        for i in range(0,N):
            if i < 50:
                output.append(reward_function(i,0,N))
            else:
                output.append(reward_function(i,1,N))
            
    return np.asarray(output)

# Model the transition probability function

def transition_probability_function(N, p, q_high, q_low, lazy = True):
    #Rows represent the current state
    #Columns represent the next state
    
    #Initiliaze the matrix
    
    P = np.zeros((N,N))
    
    #Lazy policy
    if lazy == True:
        #Same effort for different que length if lazy policy
        q_1 = q_low 
        q_2 = q_low
    else:
        #Different effort for different que length if aggressive policy
        q_1 = q_low
        q_2 = q_high
        
    for i in range(0,N): #Go through columns
        if i < 50:
            q = q_1
        else:
            q = q_2
        for j in range(0,N):#Go through rows
            if i == j == 0:
                P[i,j] = p * q + (1-p) * (1-q) + (1-p) * q
            elif i == j == 99:
                P[i,j] = p * q + p * (1-q) + (1-p) * (1-q)
            elif i == j:
                P[i,j] = p * q + (1-p) * (1-q)
            elif (j - i) == 1:
                P[i,j] = p * (1-q)            
            elif (j - i) == -1:
                P[i,j] = (1-p) * q
    return P

def value_function_bellman(N, gamma,p,q_high, q_low, lazy = True):
    if lazy == True:
        dummy = True
    else:
        dummy = False
    
    A = np.identity(N) - gamma* transition_probability_function(N, p, q_high, q_low, lazy = dummy)
    A_inv = np.linalg.inv(A)
    r = reward_policy_function(N, lazy = dummy)
    return A_inv @ r 
```

# Part 1: Policy evaluation

```python
#Set the parameters

N = 100
p = 0.5
q_high = 0.6
q_low = 0.51
gamma = 0.9
```

## Calculating the values 

```python
lazy_bellmann = value_function_bellman(N, gamma, p, q_high, q_low, lazy = True)
agg_bellmann = value_function_bellman(N, gamma, p, q_high, q_low, lazy = False)
difference = lazy_bellmann - agg_bellmann
```

## Plotting the results

```python
#Set the parameter and the style
x = np.arange(0,N)
plt.figure(figsize=(12, 5)) 

#First subplot
plt.subplot(1,2,1)
plt.plot(x, lazy_bellmann, label='lazy policy')
plt.plot(x, agg_bellmann, label='aggressive policy')
plt.title('Policy evaluation of the two strategies lazy vs. aggressive ')
plt.ylabel('value')
plt.xlabel('state')
plt.legend()


plt.subplot(1,2,2)
plt.plot(x, difference, label='difference')
plt.title('Policy evaluation difference V_lazy - V_aggressive')
plt.ylabel('difference')
plt.xlabel('state')
plt.legend()

plt.show()
```

**Comment**

As we can see, the lazy policy is better for state 50, whereas the aggressive policy is superior for state 80!

```python
def value_iteration(a, N, gamma, P_0, P_1, threshold, return_policy = False):
    #Initialize optimal policy as lazy by default
    opt_policy = np.repeat(10,N).tolist()
    V_old = np.repeat(-10, N).tolist() #Initialize high negative values
    t_start = process_time() 
    
    #Initialize empty values for V_10, V_20, V_50, V_100
    
    V_10 = V_20 = V_50 = V_100 = np.empty(N)
    
    #Run the loop
    #Initialize a delta
    delta = np.repeat(10, N)
    
    i = 0
    while (delta >= threshold).all():
        delta_new = []
        for j in range(0, N):
            #Old value
            v = V_old[j]
            V_new = -10e10
            for action in a:
                if action == 0:
                    P = P_0
                else: 
                    P = P_1
                reward = reward_function(j, action,N)
                discounted_future_reward = gamma * P[j] @ V_old
                V_new_prop = reward + discounted_future_reward
                if V_new_prop >= V_new:
                    V_new = V_new_prop
            if i == 100:
                opt_policy[j] = action
            delta = np.absolute(v-V_new)
            V_old[j] = V_new
            delta_new.append(delta)
        #Increase the run
        i += 1
        delta = np.asarray(delta_new)
        #Storing the values that are asked for in the assignment
        if i == 10:
            V_10 = np.array(V_old)
        elif i == 20:
            V_20 = np.array(V_old)
        elif i == 50:
            V_50 = np.array(V_old)
        elif i == 100:
            V_100 = np.array(V_old)
            t_stop = process_time()
    
    t_time = t_stop - t_start
    print('The time for 100 iteration is {}'.format(t_time))
    return V_10, V_20, V_50, V_100, opt_policy
        
        
```

```python
# Set the parameters for value iteration

N = 1000
gamma = 0.9
p = 0.5
q_high = 0.61
q_low = 0.51
P_0 = transition_probability_function(N,p,q_high,q_low, lazy = True)
P_1 = transition_probability_function(N,p,q_high,q_low, lazy = False)
a = [0,1]
threshold = 1e-10

V_10, V_20, V_50, V_100, opt_policy = value_iteration(a, N, gamma, P_0, P_1, threshold)
```

### Plotting the value functions

## Policy iteration

```python
Image(url= "policy_iteration_2.png", width=500, height=500)
```

**Comment**

The idea is to first implement an arbitrary value function that maps the state space to the real line. Given this, I will need to find the opimal policy. Using this optimal policy, I reevaluate the value function using my just found optimal policy. In the next step, I need to find the optimal policy again given the new value function and so on until convergence! Note that there is a type here on the right hand side in the orange box; it should be V(y) instead of V(x)!

```python
def optimal_value_policy_iteration(a, N, gamma,P_0, P_1, k, return_policy = False):    
    #Initialize optimal policy as lazy by default
    opt_policy = np.repeat(10,N).tolist()
    V_opt_policy = np.repeat(-10, N).tolist() #Initialize high negative values
    
    #Iterate through to find the optimal value
    t_start = process_time()
    for iteration in range(0, k):
        for j in range(0,N): #Go through each state
            V_x = V_opt_policy[j] #initialize a very high negative number for policy chosen at state x
            for action in a: #Go through each action
                if action == 0:
                    P = P_0
                else: 
                    P = P_1
                reward = reward_function(j, action,N)
                discounted_future_reward = gamma * P[j] @ V_opt_policy
                V_x_proposed = reward + discounted_future_reward
                if V_x_proposed >= V_x:
                    V_x = V_x_proposed
                    opt_policy[j] = action #Update the optimal action
            V_opt_policy[j] = V_x
    t_stop = process_time()
    t_end = t_stop - t_start
    print('The time for the whole iteration for {} iteration is {}'.format(k,t_end))
    
    if return_policy == False:
        return V_opt_policy
    else:
        return opt_policy
            
```

```python
# Set the parameters for policy iteration

N = 1000
gamma = 0.9
p = 0.5
q_high = 0.61
q_low = 0.51
P_0 = transition_probability_function(N,p,q_high,q_low, lazy = True)
P_1 = transition_probability_function(N,p,q_high,q_low, lazy = False)
a = [0,1]

k = [10,20,50,100]

policy_iteration = dict()

for iteration_number in k:
    policy_iteration[iteration_number] = optimal_value_policy_iteration(a, N, gamma, P_0, P_1, iteration_number)
    
```

### Comment on the runtime!

As we can see, the runtime of the policy iteration is faster than the for the value iteration. This is surprising though, as I would have expexted it to be the other way around! Yet, the difference is very small only!
