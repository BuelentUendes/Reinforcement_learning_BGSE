```python
def squared_expo_variance(x_p,x_q,p,q,l):
    kron_delta = 1 if q==p else 0
    return sigma_f*np.exp((-1/(2*l**2)) *((x_p-x_q)**2)) +sigma_n*kron_delta
```

```python
n = 100
interval = 7
```

```python
def drawing_test_points(n, interval):
    return np.linspace(-interval, interval, n)
```

```python
X_train = drawing_test_points(n, interval)
```

```python
def covariance_matrix(X_train,l):
    return [[squared_expo_variance(X_p,X_q,p,q,l) for p, X_p in enumerate(X_train)] for q, X_q in enumerate(X_train)]
```

```python
def draw_gp_functions(X_train, l,sigma_n = 0): #Add the possibility to add noise to the process, if zero the process is noise free
    return np.random.multivariate_normal(np.zeros(len(X_train)), covariance_matrix(X_train,l) + sigma_n * np.eye(len(X_train)))
```

```python
number_of_draws = 3
def draw_plot(number_of_draws, X_train):
    try:
        fig, ax = plt.subplots()
        plt.fill_between(X_train,-2*covariance_matrix(X_train,l)[0][0], 2*covariance_matrix(X_train,l)[0][0],color='grey', alpha=0.3)
        plt.xlabel('x')
        plt.ylabel('f(x)')
        plt.title('{} sample draws from a gaussian prior'.format(number_of_draws))
        for i in range(number_of_draws):
            ax.plot(X_train, draw_gp_functions(X_train,l))
        plt.show()
    except ValueError:
        print("the number of dimensions need to fit!")
```

Next, instead of implementing GP inference directly through conditioning equations, you will implement the Algorithm 2.1. Note that when doing Cholesky decomposition you need to add some very small number  to the diagonal of the covariance matrix, for numerical stability.

```python
def covariance_matrix_2(X_test, X_train,n,l):
    #n = len(X_train) +1 # To  prevent that the index is the same for the points
    n = n+1
    k = [squared_expo_variance(X_test, X_train, n, q,l) for q, X_train in enumerate(X_train)]
    return np.asarray(k)
```

```python
k = np.asarray(covariance_matrix_2(X_test, X_train, n,l))
```

```python
def prediction_log_ll_gp(X, y, sigma_n, X_test, n,l,return_marg_ll = False):
    K = covariance_matrix(X,l) + sigma_n*np.eye(len(X)) + 1e-6 * np.eye(n) #adding a small number for numerical stability
    L = cho_factor(K)
    a = cho_solve(L,y)
    
    #predictive mean
    k_x = covariance_matrix_2(X_test, X,n,l)
    f_bar = k_x.dot(a.T)
    #f_bar = np.transpose(k_x) * a
    
    #predictive variance
    
    v = cho_solve(L,k_x)
    #V_f = np.asarray(covariance_matrix([X_test]) - np.transpose(v)*v
    V_f = covariance_matrix([X_test],l) - v.dot(v.T)                
    
    #log marginal ll
    
    marg_ll = -1/2 * np.transpose(y) * a - np.trace(cholesky(K, lower = True)) - (n/2) * np.log(2 * math.pi)
    
    if return_marg_ll == False:
        return k_x, f_bar, V_f
    else: 
        return k_x, f_bar, V_f, marg_ll
```

```python
#Set the parameter values

a = 1
b = 5.1/(4*math.pi**2)
c = 5/math.pi
r = 6
s = 10
t = 1/(8*math.pi)

#Set the absolute mininimum

f_global_min = 0.397887
```

We need to evaluate the function at the points $ x_1 \in [-5,10]$ and $ x_2 \in [0,15]$. The global minimum is f(x) = 0.397887 at $x_1 = (\pi, 12.275)$, $ x_2 = (\pi, 2.275) $ and $x_3 = (9,.42478, 2475)$

### Defining the objective function 

```python
def Branin_Hoo_function_value(x_1,x_2,a = a,b = b,c = c,r = r,s = s,t = t):
    return a*((x_2-b*x_1**2 +c*x_1 -r)**2) + s*(1-t)*math.cos(x_1) + s
```

### Starting sample 

```python
n = 5
X_start_sample = np.asarray([[np.random.uniform(-5,10), np.random.uniform(0,15)] for i in range(0,n)])
```

```python
Y_start_sample = np.asarray([[Branin_Hoo_function_value(i[0],i[1])] for i in X_start_sample])
```

### Writing the acquision functions

Given the instructions of the assignment, I will define the following three acquisition functions:

```python
def UCBacquisition(X_test, muNew, stdNew, beta):
    length = X_test.shape[0]
    acq = [muNew[i][0] + beta * std_pred[i] for i in range(0,length)]
    return np.argmin(acq)

def PIacquisition(X_test, X_train, muNew, stdNew, ep):
    y = GP.predict(X_train)
    fmin = min(y)
    length = X_test.shape[0]
    Z = [((muNew[i][0]-fmin - ep)/(stdNew[i] + (1e-4))) for i in range(0, length)]
    acq = [1-stats.norm.cdf(Z[i][0]) for i in range(0,length)]
    #acq = [stats.norm.cdf(Z[i][0]) for i in range(0,length)]
    #max_ind = np.argmax(stats.norm.cdf(Z))
    max_ind = np.argmax(acq)
    return X_test[max_ind]

def EIacquisition(X_test, X_train, muNew, stdNew, ep):
    y = GP.predict(X_train)
    fmin = min(y)
    length = X_test.shape[0]
    Z = [((muNew[i][0]-fmin - ep)/(stdNew[i] + (1e-4))) for i in range(0, length)]
    acq = [(muNew[i][0]-fmin - ep)*stats.norm.cdf(Z[i][0]) + stdNew[i]*stats.norm.pdf(Z[i][0]) for i in range(0,length)]
    min_ind = np.argmin(acq)
    return X_test[min_ind]
```

### Discretinizer 

```python
def discretize(start, end, length):
    assert len(start)==len(end)
    ans = np.empty((len(start),length))
    for row, s_i, e_i in zip(range(len(start)),start, end):
         ans[row]=np.linspace(s_i, e_i, length)
    return ans.T
```

```python
x_space= discretize([-5,0], [10,15], 100)
```

```python
#Set the exploration value for beta and the UCB algorithm
beta = [0.1, 2, 10]
sim = 100
f_min = 0.397887 #This is the minimum value of the function

#This is just because of the errors I get
warnings.simplefilter('ignore')

#Create a list for all results

total_results_ucb_mae = []
total_results_ucb_cumulative = []
#fig, axs = plt.subplots(1,3, figsize = (15,5))
for b in range(0, len(beta)):
    
    #Create an empty list
    MAE = []
    MAE_cumulative = []
    
    mae_cumulative = 0
    #Take 5 random points
    n = 5
    X_start_sample = np.asarray([[np.random.uniform(-5,10), np.random.uniform(0,15)] for i in range(0,n)])
    
    Y_start_sample = np.asarray([[Branin_Hoo_function_value(i[0],i[1])] for i in X_start_sample])
    
    for i in range(100):
        
        #Fit an initialize the GP regressor
        K = RBF()
        GP = GaussianProcessRegressor(kernel = K)
    
        #Fit and optimize the parameters given the points that we have in our starting sample
        GP.fit(X_start_sample, Y_start_sample) 
        
        mu_pred, std_pred = GP.predict(x_space, return_std=True)
        next_guess = UCBacquisition(x_space,mu_pred, std_pred, beta = beta[b])
        #print(next_guess)
        
        #Evaluate the function value
        min_guess = Branin_Hoo_function_value(x_space[next_guess][0],x_space[next_guess][1])
        
        #Evaluate the loss
        mae = mean_squared_error([f_global_min], [min_guess])
        mae_cumulative += mae
        #Append it to the list
        MAE.append(mae)
        MAE_cumulative.append(mae_cumulative)
        
        #Adding the explored point to the list
        X_start_sample = np.vstack((X_start_sample, x_space[next_guess]))
        Y_start_sample = np.vstack((Y_start_sample, min_guess))
        
    total_results_ucb_mae.append(MAE)
    total_results_ucb_cumulative.append(MAE_cumulative)
```
