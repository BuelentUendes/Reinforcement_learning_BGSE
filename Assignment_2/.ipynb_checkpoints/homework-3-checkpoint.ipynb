```python
#Use the function (2.31) from the book; SE Kernel with noise:
def SE_kernel(xp, xq, sigma_f, sigma_n, l):
    
    #Kronecker delta, Î´ pq = 1 iff p = q and 0 otherwise (Symbols and Notation section from the book)
    if xp == xq:
        kronecker_delta = 1
    else:
        kronecker_delta = 0
    kernel = (sigma_f**2)*np.e**((-1/(2*l**2))*((xp - xq)**2)) + (sigma_n**2 * kronecker_delta)  
    return kernel


#Create the matrix
def kernel_cov_matrix(X,  sigma_f, sigma_n, l):
    
    #initiate empty covariance matrix
    cov_matrix = np.empty(shape=(len(X), len(X)))

    #fill the matrix element by element using the SE kernel function
    for i in range(len(X)):
        for j in range(len(X)):
            cov_matrix[i,j] = SE_kernel(X[i], X[j], sigma_f, sigma_n, l)

    return cov_matrix
```

### Part b:

```python
def GP_regression(X, y, X_star, sigma_n, sigma_f, l):
    
    K = kernel_cov_matrix(X, sigma_f, sigma_n, l)
        
    f_mean = []
    f_var = []
    
    #add small number e to diagonal of covariance matrix for stability
    K_copy = np.copy(K)
    e = 1e-04
    np.fill_diagonal(K_copy, K_copy.diagonal() + e)
    
    #Cholesky decomposition:
    a = K_copy + (sigma_n**2)*(np.identity(len(K_copy)))
    L = cholesky(a, lower=True)
    alpha = cho_solve((L,True),y)
    
    
    #find the mean and var for each test point
    for i in X_star:
        
        #Acc. to book: In the case that there is only one test point xâ we write k(xâ)=kâ 
        #to denote the vector of covariances between the test point and the n training points.
        covariance = [SE_kernel(i, train_i, sigma_f, sigma_n, l) for train_i in X]
        
        mean = np.array(covariance).dot(alpha)
        
        v = np.dot(inv(L),np.array(covariance))
        
        var = (SE_kernel(i, i, sigma_f, sigma_n, l))- v.dot(v)
        
        
        f_mean.append(mean)
        f_var.append(var)
    
    
    #log_p_y_x = (-0.5*(np.dot(y.transpose(), alpha)))- (np.trace(np.log(L))) - (0.5*len(y)*(np.log(2*np.pi)))
    
    
    return f_mean, f_var, y
```

### Part d:

Function values f* (corresponding to test inputs X*) can be sampled from the
joint posterior distribution by evaluating the mean and covariance matrix from
eq. (2.19) and generating samples according to the method described in section
A.2.

```python
#2.5-a
l = 1
sigma_f = 1
sigma_n = 0.1
N = 100
start = -7
end = 7
X_test = np.linspace(-7, 7, N)

#training data
X= np.random.uniform(low=-7, high=7, size=20)
cov_training = kernel_cov_matrix(X, sigma_f, sigma_n, l)
y = np.random.multivariate_normal(mean=np.zeros(len(training_points)), cov=cov_training)


f_mean, f_var, y = GP_regression(X, y, X_test, sigma_n, sigma_f, l)

fig, ax = plt.subplots()

ub = np.array(f_mean) + 2*np.sqrt(np.array(f_var))
lb = np.array(f_mean) - 2*np.sqrt(np.array(f_var))

ax.plot(X_test, f_mean, color = 'red', label='Mean')
ax.plot(X_test, ub, color = 'lightblue')
ax.plot(X_test, lb, color = 'lightblue')
ax.fill_between(X_test, ub, lb, color = 'lightblue')

# Sample from posterior distribution:
output = []
for i in range(len(X_test)):
    output.append(np.random.normal(f_mean[i], f_var[i]))

plt.plot(X_test, output, color = 'green', label='One sample')
plt.plot(X, y, 'x', color='black', label='Training Points')
plt.legend(loc="upper right")
ax.set(title='Figure 2.5-a')
```

## Exercise 2:

### Branin Function:

```python
def obj_func(x1, x2):
    a = 1
    b = (5.1)/(4*(np.pi**2))
    c = 5/np.pi
    r = 6
    s = 10
    t = 1/(8*np.pi)
    
    #Put minus sign because we want to find global min:
    f =  -( a*((x2-b*(x1**2)+c*x1-r)**2)+s*(1-t)*(np.cos(x1)) + s )
    return f
```

### Data:

```python
#This function discretizes the 2-D space into a finite element grid
def discretize(start, end, length):
    ans = np.empty((len(start),length))
    for row, s_i, e_i in zip(range(len(start)),start, end):
         ans[row]=np.linspace(s_i, e_i, length)
    return ans.T

def data_creator(start, end, length, seed):
    #set seed so that I generate same data every time
    sp.random.seed(seed)
    
    #start with 5 random points uniformly chosen from the space x1 â [â5, 10] and x2 â [0, 15]
    x1 = sp.random.uniform(-5,10,5)
    x2 = sp.random.uniform(0,15,5)
    X_train = np.array([x1,x2]).T
    
    y_true = [obj_func(x[0], x[1]) for x in X_train]
    
    X_test = discretize(start, end, length)
    
    return X_train, y_true, X_test
```

### Acquisition Functions:

```python
def UCBacquisition(X_test, muNew, stdNew, beta):
    max_ind = np.argmax(np.array(muNew) + beta*np.array(stdNew))
    return X_test[max_ind]

def PIacquisition(X_test, X_train, muNew, stdNew, ep):
    y = GP.predict(X_train)
    #print(y)
    fmax = max(y)
    #print(fmax)
    Z = (muNew - fmax - ep)/(stdNew)
    max_ind = np.argmin(stats.norm.cdf(Z))
    return X_test[max_ind]

def EIacquisition(X_test, X_train, muNew, stdNew, ep):
    y = GP.predict(X_train)
    fmax = max(y)
    Z = (muNew - fmax - ep)/(stdNew)
    max_ind = np.argmin((muNew - fmax - ep)*stats.norm.cdf(Z) + stdNew*stats.norm.pdf(Z))
    return X_test[max_ind]
```

#### UCB Simulations:

```python
#I use the gp regresspr implemented in the sklearn package
GP = GaussianProcessRegressor()

#UCB parameters:
params = [0.01, 5, 20]

#This is just because of the errors I get
warnings.simplefilter('ignore')

fig, axs = plt.subplots(1,3,figsize=(15,5))

#iterate over each parameter
for i in range(len(params)):
    
    squared_err = []
    
    #generate data for each parameter, the data has random seed set for reproducability
    X_train, y_true, X_test = data_creator([-5, 0], [10, 15], 100, 10)

    #iterate 100 times
    for j in range(100):
        
        #fit guassian process regressor
        GP.fit(X_train,y_true)

        #predict the new mean and standard variation using guassian process regressor
        muNew, stdNew = GP.predict(X_test, return_std=True) 

        #find the new X and respective y using the parameter
        XNew = UCBacquisition(X_test, muNew, stdNew, params[i])
        yNew = obj_func(XNew[0], XNew[1])
        
        #compute the squared error and append it to scorer of squared error
        se = (yNew - 0.397887)**2
        squared_err.append(se)
        
        #append the new X and y to the original
        X_train = np.vstack((X_train, [XNew]))
        y_true = np.append(y_true, yNew)
        
    axs[i].plot([n for n in range(100)],squared_err)
    axs[i].title.set_text('UCB with parameter = {}'.format(params[i]))
    axs[i].set_xlabel('Iterations')
    axs[i].set_ylabel('Squared error difference')
fig.tight_layout()
```

Even though all the figures look very similar, one can notice that having e = 0.01 achieves a squared error = 0 faster than e = 20.
